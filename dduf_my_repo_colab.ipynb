```python
model_id = "CompVis/stable-diffusion-v1-4"
user_name = "reach-vb"
dduf_file_name = model_id.split("/")[-1] + ".dduf"
upload_repo_id = f"{user_name}/{dduf_file_name}"
```

```python
import os

os.environ["HF_HUB_ENABLE_HF_TRANSFER"] = "1"
```

```python
from huggingface_hub import snapshot_download
snapshot_download(repo_id=model_id, local_dir="models", revision="main")
```

```python
from huggingface_hub import export_folder_as_dduf
export_folder_as_dduf(dduf_file_name, folder_path="models")
```

```python
from huggingface_hub import create_repo
create_repo(repo_id=upload_repo_id)
```

```python
from huggingface_hub import upload_file
upload_file(path_or_fileobj=dduf_file_name, repo_id=upload_repo_id, path_in_repo=dduf_file_name)
```
