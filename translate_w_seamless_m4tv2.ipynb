# Seamless M4T v2 Large by Meta AI

You can translate longer chunks of audio via the pipeline API in transformers. This helps bypass the current limitation of the seamlessM4T models only accepting upto 60 seconds of audio.

This colab notebook serves as quite a simple demostration on how to use the SeamlessM4T checkpoints.

## Initialise the SeamlessM4T v2 checkpoints

```python
import torch
from transformers import pipeline

translator = pipeline("automatic-speech-recognition",
                      "facebook/seamless-m4t-v2-large",
                      torch_dtype=torch.float16,
                      device="cuda:0")
```

## Translate! 🚀

```python
translator("https://huggingface.co/datasets/reach-vb/random-audios/resolve/main/ted_60.wav",
           chunk_length_s=30,
           generate_kwargs={"tgt_lang": "fra"})
```

You can swap the `tgt_lang` to swap and get translations in different languages.

```python
translator("https://huggingface.co/datasets/reach-vb/random-audios/resolve/main/ted_60.wav",
           chunk_length_s=30,
           generate_kwargs={"tgt_lang": "ita"})
```

```python
translator("https://huggingface.co/datasets/reach-vb/random-audios/resolve/main/ted_60.wav",
           chunk_length_s=30,
           generate_kwargs={"tgt_lang": "deu"})
```
