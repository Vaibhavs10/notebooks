## Initialise the Text Generation pipeline

P.S. Make sure to accept the terms and conditions from the model page [here](https://huggingface.co/mistralai/mathstral-7B-v0.1)

```python
from transformers import pipeline
import torch

model_id = "mistralai/mathstral-7B-v0.1"

pipe = pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)
```

## Define the prompt similar to messages API

You can pretty much define any query, question, discussion that you want the LLM to process below.

```python
messages = [
    {"role": "user", "content": "Albert likes to surf every week. Each surfing session lasts for 4 hours and costs $20 per hour. How much would Albert spend in 5 weeks?"},
]
```

## Pass the outputs to the pipeline

```python
outputs = pipe(
    messages,
    max_new_tokens=256,
    do_sample=False,
)
```

## Generate response

```python
assistant_response = outputs[0]["generated_text"][-1]["content"]
print(assistant_response)
```
